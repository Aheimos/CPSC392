# Visualization
[This](https://pandas.pydata.org/pandas-docs/stable/visualization.html) is a great reference from pandas

```python
import matplotlib.pyplot as plt
import pandas
import numpy as np
```

### Basic line plot

```python
# consistent interval for x-axis
x = [x*0.1 for x in range(100)]
# function of x for y-axis
y = np.sin(x)
# create line plot
plt.plot(x, y)
# show line plot
plt.show()
```

### Scatter plot

```python
# first variable
x = 20 * np.random.randn(1000) + 100
# second variable
y = x + (10 * np.random.randn(1000) + 50)
# create scatter plot
plt.scatter(x, y)
# show line plot
plt.show()
```

### Bar plot

```python
objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
y_pos = np.arange(len(objects))
performance = [10,8,6,4,2,1]
 
fig = plt.figure(figsize=(15,6))
plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Usage')
plt.title('Programming language usage')
 
plt.show()
```

### Histogram

```python
# random numbers drawn from a Gaussian distribution
x = np.random.randn(1000)
fig = plt.figure(figsize=(15,6))
# create histogram plot
plt.hist(x)
plt.show()
```

### Box and whisker

```python
# random numbers drawn from a Gaussian distribution
x = [np.random.randn(1000), 5 * np.random.randn(1000), 10 * np.random.randn(1000)]
fig = plt.figure(figsize=(15,6))
# create box and whisker plot
plt.boxplot(x)
plt.show()
```

### Pandas Visualizations

```python
# load data
from pandas.tools.plotting import scatter_matrix
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = pandas.read_csv(url, names=names)
data.head()
```

```python
# Univariate Histograms
data.hist(figsize=[15, 15])
plt.show()
```

```python
# Univariate Density Plots
data.plot(kind='density', subplots=True, layout=(3,3), sharex=False,figsize=[15, 15])
plt.show()
```

```python
# Box and Whisker Plots
data.plot(kind='box', subplots=True, layout=(3,3), sharex=False, sharey=False,figsize=[15, 15])
plt.show()
```

```python
# Correction Matrix Plot
correlations = data.corr()
# plot correlation matrix
fig = plt.figure(figsize=(15,6))
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,9,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
```

Like the Correlation Matrix Plot, the scatterplot matrix is symmetrical. This is useful to look at the pair-wise relationships from different perspectives. Because there is little point oi drawing a scatterplot of each variable with itself, the diagonal shows histograms of each attribute.
