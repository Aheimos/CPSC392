```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
```

```python
x = np.linspace(-5,5,11).reshape(1,11).T
y = x + np.random.normal(size=x.shape)
```

```python
plt.grid()
plt.scatter(x,y)
```

```python
a = np.matmul(x.T, x)
b = np.linalg.inv(a)
c = np.matmul(b, x.T)
theta = np.matmul(c, y)
```

```python
plt.grid()
plt.plot(x, theta*x, color='r')
plt.scatter(x,y)
```
